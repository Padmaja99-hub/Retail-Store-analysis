```python
import pandas as pd
df = pd.read_csv('orders.csv', na_values= ['Not Available','unknown'])
df.head()
#df['Ship Mode'].unique()
```

```python
#df.rename(columns={'City': 'city'}) - to remane using json
df.columns= df.columns.str.lower()
df.columns= df.columns.str.replace(' ','_')
df.head()
#renamed all column names with lowercase and _ instead of space in between words
```

```python
#derive new columns discount, sales price and profit
df['discount']=df['cost_price']*df['discount_percent']*0.01
df['sales_price']=df['list_price']-df['discount']
df['profit']=df['sales_price']-df['cost_price']
df.head()
```

```python
df['order_date']= pd.to_datetime(df['order_date'], format='%Y-%m-%d')
#converting obj type to datetime 
```
